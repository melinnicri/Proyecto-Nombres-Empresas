```python
# Este notebook se realizó para comparar los datos extraídos desde el escrapeo
# Librerías para comparar resultados
import pandas as pd
```

```python
# Cargar datos de los nombres de las empresas antes del escrapeo
df_original = pd.read_csv("../proyecto_correccion/empresas_limpias_corregidas_final.csv")
```

```python
# Cargar datos del escrapeo resultante
df_scraping = pd.read_csv("contacto_empresas_es.csv")
```

```python
# Unir los dataframes por la columna 'CIF'
df_comparado = pd.merge(df_original, df_scraping, on="CIF", how="left")
```

```python
# Calcular cobertura por campo
total_empresas = len(df_comparado)

cobertura_url = (df_comparado["URL"] != "No encontrado").sum()
cobertura_direccion = (df_comparado["Dirección"] != "No encontrado").sum()
cobertura_telefono = (df_comparado["Teléfono"] != "No encontrado").sum()
cobertura_email = (df_comparado["Email"] != "No encontrado").sum()

print(f"Cobertura url: {cobertura_url}/{total_empresas} ({cobertura_url/total_empresas:.1%})")
print(f"Cobertura dirección: {cobertura_direccion}/{total_empresas} ({cobertura_direccion/total_empresas:.1%})")
print(f"Cobertura teléfono: {cobertura_telefono}/{total_empresas} ({cobertura_telefono/total_empresas:.1%})")
print(f"Cobertura email: {cobertura_email}/{total_empresas} ({cobertura_email/total_empresas:.1%})")
```

```python
# Empresas con todos los datos disponibles (aunque no necesariamente correctos)

# Normalizar valores nulos o marcadores de error
valores_invalidos = ["No encontrado", "No encontrada", "Error", "", None]

# Crear una columna booleana que indique si todos los campos están llenos
df_comparado["completo"] = ~df_comparado[["URL", "Dirección", "Teléfono", "Email"]].isin(valores_invalidos).any(axis=1)

# Contar cuántas empresas tienen todos los campos llenos
total_completas = df_comparado["completo"].sum()
total_empresas = len(df_comparado)

print(f"Empresas con todos los campos llenos: {total_completas}/{total_empresas} ({total_completas/total_empresas:.1%})")
```

```python
# Empresas con los datos completos
empresas_completas = df_comparado[df_comparado["completo"]]
empresas_completas[["Empresa", "URL", "Dirección", "Teléfono", "Email"]]

```

```python
empresas_completas[["Empresa", "URL", "Dirección", "Teléfono", "Email"]].to_csv("empresas_completas.csv", index=False)
print("Datos de empresas con todos los campos guardados en empresas_completas.csv")
```

```python
# Cálculo de cobertura y rendimiento general, usando la columna "completo"
total_empresas = len(df_comparado)
total_completas = df_comparado["completo"].sum()

cobertura = (total_completas / total_empresas) * 100
print(f"Cobertura de extracción: {total_completas}/{total_empresas} ({cobertura:.1f}%)")
```

```python
#¿Cómo costearlo?
#Si quisiera estimar el costo por empresa útil (19 empresas), considerando todo el trabajo:

#python
horas_totales = 36
empresas_utiles = 19
costo_por_empresa = horas_totales / empresas_utiles
print(f"Costo estimado por empresa útil: {costo_por_empresa:.2f} horas")
#Resultado: ≈ 1.89 horas por empresa útil, incluyendo todo el flujo.
```

```python
# Definir qué cuenta como error
valores_invalidos = ["Error", "", None]
campos_clave = ["URL", "Dirección", "Teléfono", "Email"]
```

```python
# Contar errores por fila
df_comparado["errores"] = df_comparado[campos_clave].apply(lambda fila: sum(valor in valores_invalidos or pd.isna(valor) for valor in fila), axis=1)
```

```python
# Filtrar empresas con ≥3 errores
filas_con_errores = df_comparado[df_comparado["errores"] >= 3]
print(f"Empresas con al menos 3 errores: {len(filas_con_errores)}")
```

```python
# Ahora buscando con nada de datos
valores_invalidos = ["Error", "", "No encontrado", "No encontrada", None]
campos_clave = ["URL", "Dirección", "Teléfono", "Email"]
```

```python
def sin_datos(row):
    return all(row[campo] in valores_invalidos or pd.isna(row[campo]) for campo in campos_clave)

df_comparado["sin_datos"] = df_comparado.apply(sin_datos, axis=1)
```

```python
# Filtrar empresas sin ningún dato útil
empresas_sin_datos = df_comparado[df_comparado["sin_datos"]]
print(f"Empresas sin ningún dato útil: {len(empresas_sin_datos)}")
```

```python
# Resumen
total = len(df_comparado)
con_errores = len(filas_con_errores)
sin_datos = len(empresas_sin_datos)

print("Resumen de cobertura:")
print(f"• Total de empresas procesadas: {total}")
print(f"• Cobertura de extracción: {total_completas}/{total_empresas} ({cobertura:.1f}%)")
print(f"• Cobertura url: {cobertura_url}/{total_empresas} ({cobertura_url/total_empresas:.1%})")
print(f"• Cobertura dirección: {cobertura_direccion}/{total_empresas} ({cobertura_direccion/total_empresas:.1%})")
print(f"• Cobertura teléfono: {cobertura_telefono}/{total_empresas} ({cobertura_telefono/total_empresas:.1%})")
print(f"• Cobertura email: {cobertura_email}/{total_empresas} ({cobertura_email/total_empresas:.1%})")
print(f"• Costo estimado por empresa útil: {costo_por_empresa:.2f} horas")
print(f"• Empresas con al menos 3 errores: {con_errores}")
print(f"• Empresas sin ningún dato útil: {sin_datos}")
print(f"• Costo estimado por empresa útil: {costo_por_empresa:.2f} horas")
```
